```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Data from the table provided by the user
data = {
    "Enterprise": [255235, 208131, 193666, 195586, 53874, 226242],
    "School": [74236, 45837, 36895, 45966, 15826, 55844],
    "Personal": [11724, 14915, 14740, 15012, 7193, 11550],
    "Labs": [14172, 9369, 9028, 10292, 2285, 11630],
    "Others": [3569, 2867, 3059, 2859, 1101, 2939],
    "Government": [373, 307, 337, 320, 95, 386]
}

# Converting the data into a pandas DataFrame
df = pd.DataFrame(data)
df.index = ['Dataset1', 'Dataset2', 'Dataset3', 'Dataset4', 'Dataset5', 'Dataset6']

# Plotting grouped bar chart
ax = df.plot(kind='bar', stacked=False, figsize=(12, 8))
ax.set_xlabel('Dataset')
ax.set_ylabel('Applicant')
ax.set_title('Applicant Composition Anlysis')
plt.xticks(rotation=0)
plt.tight_layout()

# Show the plot

plt.savefig('../assets/composition_analysis.png')
plt.show()

# Now let's create a heatmap for the same data
plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, fmt="d", linewidths=.5, cmap='Oranges')
plt.title('Hot Map of Applicant Composition Anlysis')
plt.ylabel('Dataset')
plt.xlabel('Applicant Type')
plt.tight_layout()

# Show the heatmap

plt.savefig('../assets/hot_map_composition_analysis.png')
plt.show()
```
